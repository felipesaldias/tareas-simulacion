```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from IPython.display import display
import ipywidgets as widgets
from IPython.core.pylabtools import figsize

def ecdf(data):
    """ Calcula la distribución empírica acumulada"""
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)
datos = stats.binom.rvs(10,0.5,size=1000)
```

```python
## Simularemos datos de una distribución NORMAL y luego graficaremos su
## función de distribución empírica y la distribución teórica
mu=5
sigma=10
datos = stats.norm.rvs(loc=mu, scale=sigma, size=1000)

fig, axs = plt.subplots(1, 2,figsize=(10, 4), dpi=80)

axs[0].hist(datos, edgecolor = 'black',
         bins = int(70/5))
ax1 = axs[0].twinx()
vec = range(-30,40,2)
##
dvec = []
for i in vec:
    dvec.append(stats.norm.pdf(x=i, loc=mu, scale=sigma))
ax1.plot(vec,dvec,color='red')

x,cum_norm = ecdf(datos)
axs[1].bar(x,cum_norm)
## probabilidad teórica
ax2 = axs[1].twinx()
vec = range(-30,40,2)
##
pvec = []
for i in vec:
    pvec.append(stats.norm.cdf(x=i, loc=mu, scale=sigma))
ax2.plot(vec,pvec,color='red')
plt.draw()
```

```python
stats.norm.pdf(x=1, loc=1, scale=1)
```

```python
## Calculemos ahora los estimadores máximo verosímiles y veamos como quedan las curvas
mu0,sigma0 = stats.norm.fit(datos)

fig, axs = plt.subplots(1, 2,figsize=(10, 4), dpi=80)

axs[0].hist(datos, edgecolor = 'black',
         bins = int(70/5))
ax1 = axs[0].twinx()
vec = range(-30,40,2)
##
dvec0 = []
for i in vec:
    dvec0.append(stats.norm.pdf(x=i, loc=mu0, scale=sigma0))
    
ax1.plot(vec,dvec0,color='black')
ax1.plot(vec,dvec,color='red')

axs[1].bar(x,cum_norm)
## probabilidad teórica
ax2 = axs[1].twinx()
vec = range(-30,40,2)
##
pvec0 = []
for i in vec:
    pvec0.append(stats.norm.cdf(x=i, loc=mu0, scale=sigma0))
ax2.plot(vec,pvec0,color='black')   
ax2.plot(vec,pvec,color='red')
plt.draw()
```

```python
## Utilicemos ahora el test de K-S para probar el ajuste de datos
## respecto de la distribución teórica?
stats.kstest(datos, 'norm')
```

```python
## Utilicemos ahora el test de K-S para probar el ajuste de datos
## respecto de la distribución teórica?
stats.kstest(datos, 'norm', args=(mu, sigma))
```

```python
## Respecto de la distribución con los parámetros MV?
stats.kstest(datos, 'norm', args=(mu0, sigma0))
```

```python
## Comparando con una nueva muestra extraida de la distribución con los parámetros MV?
datos2 = stats.norm.rvs(loc=mu0, scale= sigma0,size=1000)

stats.kstest(datos, datos2)
```

```python
##Distribución bernoulli(discreta)
p = 0.6
n = 1000
datos = stats.bernoulli.rvs(p, size=n, random_state=2)
```

Para poder realizar el ajuste, ya que no existe un "fit" que nos permita ajustar una distribución discreta, utilizamos el maximum likelihood para bernoulli, que por lo que vimos en clases era igual a la frecuencia empírica: 

$$ \hat{p}= \frac{\sum X_i}{n}$$

```python
p_0=(n-datos.sum())/n
p_1=datos.sum()/n
```

```python
posible_value = [0, 1]
puntual_max_likelihood_prob = [p_0, p_1]
puntual_theoric_prob = [1-p, p]

cumulative_max_likelihood_prob = [p_0, p_0+p_1]
cumulative_theoric_prob = [(1-p), p+(1-p)] 
```

```python
fig, axs = plt.subplots(1, 2,figsize=(10, 4), dpi=80)
axs[0].scatter(posible_value, puntual_max_likelihood_prob, color='red')
axs[0].scatter(posible_value, puntual_theoric_prob, color='blue')
axs[0].set_ylim(0,1)
axs[0].set_xticks([0,1])
#ax1 = axs[0].twinx()


x,cum_norm = ecdf(datos)
axs[1].plot(posible_value, cumulative_max_likelihood_prob, color='red')
axs[1].plot(posible_value, cumulative_theoric_prob, color='blue')
axs[1].set_xlim([0,1])
axs[1].set_ylim(0,1)
axs[1].set_xticks([0,1])
```

## **Ahora vamos a hacer la comparacion para las aerolineas**


Los graficos que se muestran abajo contienen los histogramas, las funciones de densidad y una linea roja indicando los 0 minutos de retreaso ni adelanto. Esta se obtiene a partir del valor mínimo que es restado en cada una de los grupos de datos para desplazarlos a valores positivos. por lo tanto los que se encuentren ahi seran vuelos que llegaron a la hora.

```python
#encontramos las diferentes aerolineas
airlines = vuelos['name'].unique()
airlines
```

```python
#airlines = np.delete(airlines, np.where(airlines == 'Southwest Airlines Co.'))
#print(airlines)
```

```python
qty_airlines = len(airlines)
fig, axs = plt.subplots(qty_airlines, 1, figsize=(8, 40))#, dpi=80)
fig.tight_layout(pad=2)

for i, x in enumerate(airlines):
    
    by_airline=vuelos[vuelos['name']==x]
   
    datos1 = by_airline['arr_delay']
    datosMin = datos1.min()
    datos1 = datos1 - datosMin
    axs[i].hist(datos1, color = 'lightblue', edgecolor = 'black',
         bins = int(180/5))
    
    axs[i].axvline(-datosMin, color='red')
    
    ax2 = axs[i].twinx()
    
    s, loc, scale = stats.lognorm.fit(datos1)
    axs[i].set_title(x+" min:{} s: {} loc: {}, scale: {}".format(datosMin,round(s,3),round(loc,3),round(scale,3)))
    
    vec = list(range(1,181))
    
    pvec = []
    for i in vec:
        pvec.append(stats.lognorm.pdf(x=i,s=s, loc=loc, scale=scale))
    ax2.plot(vec,pvec)
    
 

    
```

Si bien todos comparten un comportamiento similar, en donde los gráficos quedan visiblemente asimetricos positivos por lo que se infiere que en este rubro la tendencia es llegar antes de la hora prevista. 14 de las 16 aerolineas se ajustaron de una buena manera a la distribucion LogNormal, la ultima claramente es por falta de datos que no se ajusta correcamente, y el caso de south-west airlines no tiene explicacion para nosotros.
